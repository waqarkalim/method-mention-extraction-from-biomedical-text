```python
import pickle
import os
```

```python
with open('complex_rules_iob_input.pkl', 'rb') as f:
    dt_list = pickle.load(f)
```

```python
word_list = list(set([x[0].lower() for y in dt_list for x in y]))
```

```python
# This cell makes a dictionary of only the words in our dataset rather than having every word in the BioWordVec file, this makes it faster to do future processing later on

word_embeddings_dictionary = dict()

folder_name = '.'
file = 'BioWord2Vec_standard.w2v'

path = os.path.join(folder_name, file)
i = 0
with open(path, 'r') as f:
    for embedding_text in f:
        word = embedding_text.split(" ", 1)[0]
        embedding = [float(x) for x in (embedding_text.split(" ", 1)[1].strip('\n')).split(" ")]
        if (len(embedding) == 200):
#             print(word, embedding, '\n')
            if word in word_list:
                word_embeddings_dictionary[word] = embedding
        if (i % 100000 == 0):
            print(i)
#             print(word_embeddings_dictionary)
#             print('\n\n')
        i+=1

with open('word_embeddings_dict.pkl', 'wb+') as f:
    pickle.dump(word_embeddings_dictionary, f)
```
